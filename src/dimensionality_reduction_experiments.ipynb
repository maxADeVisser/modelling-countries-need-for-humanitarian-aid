```python
import matplotlib.pyplot as plt
import numpy as np
```

# Curse of Dimensionality

```python
def curse_of_dimensionality(dimension_range: tuple, point_range: tuple, iterations: int = 5):
    """Visualises the curse of dimensionality"""
    
    def dist(p1: list, p2: list):
        """Calculates the Euclidean distance between two points in n-dimensional space (the length of the points)"""
        return np.sqrt(np.sum([(p1[i] - p2[i])**2 for i in range(len(p1))]))

    dimensions = np.arange(dimension_range[0], dimension_range[1])
    
    for i in range(iterations):
        p1 = [] 
        p2 = [] 
        distances = []

        for i in range(len(dimensions)):
            p1.append(np.random.randint(point_range[0], point_range[1]))
            p2.append(np.random.randint(point_range[0], point_range[1]))
            distances.append(dist(p1, p2))
            
        plt.plot(dimensions, distances)
    
    plt.xlabel("# Dimensions")
    plt.ylabel("Euclidean dist between Point 1 and 2")
    plt.title(f"Curse of Dimensionality")
    plt.text(0.8, 0.382, f"The point axis-values\nare fixed in interval:\n{point_range[0]} to {point_range[1]}", ha='center', va='center', transform=plt.gca().transAxes)
    plt.grid()
    plt.show()
    
curse_of_dimensionality(dimension_range=(1, 1000), 
                        point_range=(1, 10),
                        iterations=10)
```
